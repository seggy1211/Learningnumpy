Student analysis and vizualization

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
"""Getting the data"""
np.random.seed(10)
Info = np.random.randint(60, 101, size = (50, 8))
Info
```

```python
""" Mean, median, std, mean for every student"""
mean_s = np.mean(Info,  axis=0)
median_s = np.median(Info)
std_s = np.std(Info)
mean_each = np.mean(Info, axis=1)
mean_s, median_s, std_s, mean_each
```

```python
""" Getting 3 highest scores """
# highest = np.sort(mean_each)[-3:]
# print (highest)
top_students_indices = np.argsort(-mean_each)[:3]  # Indices of top 3 students
# top_students_indices = np.argsort(mean_each) getting all indices according to from lowest to highest scores
print(top_students_indices)
top_students_grades = Info[top_students_indices]

print("Top Performing Students' Grades:\n", top_students_grades)
```

```python
subjects = ['Math', 'Science', 'History', 'English', 'Art', 'Geography', 'Physics', 'Chemistry']
plt.figure(figsize=(10, 6))
plt.bar(subjects, mean_s, color='skyblue')
plt.xlabel('Subjects')
plt.ylabel('Mean Grades')
plt.title('Average Grades per Subject')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
""" Subject correlation """
subject_correlation = np.corrcoef(Info, rowvar=False)
print('Correlation Between', subject_correlation)
```

```python
plt.figure(figsize=(8, 6))
plt.imshow(subject_correlation, cmap='viridis', interpolation='nearest')
plt.colorbar()
plt.title('Correlation Between Subjects')
plt.xticks(np.arange(len(subjects)), subjects, rotation=45)
plt.yticks(np.arange(len(subjects)), subjects)
plt.tight_layout()
plt.show()
```
